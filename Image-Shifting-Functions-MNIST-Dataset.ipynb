# Title:
Image Shifting with MNIST Dataset 

```python
# Import necessary libraries
import numpy as np
import pandas as pd
from scipy.ndimage import shift
from sklearn.datasets import fetch_openml
import matplotlib as mpl
import matplotlib.pyplot as plt
import scipy.ndimage.interpolation as ndi_int
import warnings

# Ignore warnings
warnings.filterwarnings('ignore')
```

```python
# Load the Data from the MNIST dataset
mnist = fetch_openml('mnist_784', version=1)

# Check the type of the mnist variable
type(mnist)
```

```python
# Separate the data and target variables
X, y = mnist["data"], mnist["target"].astype(int)

# Check the types of X and y
type(X), type(y)
```

```python
# Convert X into a pandas DataFrame and y into a pandas Series
X = pd.DataFrame(X)
y = pd.Series(y)
```

```python
# Check the types of X and y after conversion
type(X), type(y)
```

```python
# Display the first few rows of X
X.head()
```

```python
# Display the first few rows of y
y.head()
```

```python
# Select a sample digit from the dataset
digit = X.iloc[0,:].values

# Reshape the digit to match the dimensions of the original image (28x28 pixels)
digit_img = digit.reshape(28,28)

# Display the original digit image
plt.imshow(digit_img, cmap = mpl.cm.binary, interpolation="nearest")
plt.axis("off")
plt.show()
```

# Shift Image Up

```python
# Funtion to move the image
def shift_image(digit_img,x,y):
    shifted_image = ndi_int.shift(digit_img,[x,y])
    fig, axes = plt.subplots(1, 2, figsize=(14, 14))
    axes[0].imshow(digit_img, cmap=mpl.cm.binary)
    axes[0].set_title('Original Image')
    axes[0].axis('off')
    axes[1].imshow(shifted_image, cmap=mpl.cm.binary)
    axes[1].set_title('Shifted Image')
    axes[1].axis('off')
    plt.show()
```

```python
# Testing the function by shifting the digit image 5 pixels up
shift_image(digit_img,-5,0)
```

```python
# Shift Image Up 1 pixel
shift_image(digit_img,-1,0)
```

# Shift Image Down

```python
# Testing the function by shifting the digit image 5 pixels down
shift_image(digit_img,5,0)
```

```python
# Shift the digit image 1 pixel down
shift_image(digit_img,1,0)
```

# Shift Image Right

```python
# Testing the function by shifting the digit image 5 pixels to the right
shift_image(digit_img,0,5)
```

```python
# Shift the digit image 1 pixel to the right
shift_image(digit_img,0,1)
```

# Shift Image Left

```python
# Testing the function by shifting the digit image 5 pixels to the left
shift_image(digit_img,0,-5)
```

```python
# Testing the function by shifting the digit image 1 pixel to the left
shift_image(digit_img,0,-1)
```
